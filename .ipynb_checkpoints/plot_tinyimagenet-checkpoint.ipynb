```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# With Dropout

## VGG13

```python
# vgg13_full_vanilla_2
with open('./data_tinyimagenet/vgg13_full_vanilla_3.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])
train_1_top5 = np.array(data['train_top5']) * 100
val_1_top5 = np.array(data['val_top5']) * 100

# vgg13_full_lc01_frac8
with open('./data_tinyimagenet/vgg13_full_lc05_frac4_2.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])
train_2_top5 = np.array(data['train_top5']) * 100
val_2_top5 = np.array(data['val_top5']) * 100


plt.plot(train_acc1, label='train vanilla')
plt.plot(val_acc1, label='val vanilla')

plt.plot(train_acc2, label='train lc01')
plt.plot(val_acc2, label='val lc01')


plt.legend()
plt.show()

plt.plot(train_1_top5, label='train vanilla')
plt.plot(val_1_top5, label='val vanilla')

plt.plot(train_2_top5, label='train lc01')
plt.plot(val_2_top5, label='val lc01')


plt.legend()
plt.show()

print("max val acc vanilla: ", np.max(val_acc1))
print("max val acc lc01: ", np.max(val_acc2))

print("max val top5 vanilla: ", np.max(val_1_top5))
print("max val top5 lc01: ", np.max(val_2_top5))
```
