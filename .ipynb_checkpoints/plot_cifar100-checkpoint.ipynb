```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# Full Dataset VGG Dropout

### VGG 13

```python
# vgg13_full_vanilla
with open('./data_cifar100_dropout/vgg13_full_vanilla_3.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])
train_1_top5 = np.array(data['train_top5']) * 100
val_1_top5 = np.array(data['val_top5']) * 100

# vgg13_full_lc1
with open('./data_cifar100_dropout/vgg13_full_lc05_frac4_3.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])
train_2_top5 = np.array(data['train_top5']) * 100
val_2_top5 = np.array(data['val_top5']) * 100

plt.plot(train_acc1, label='train_acc1 (vgg13_full)')
plt.plot(val_acc1, label='val_acc1 (vgg13_full)')
plt.plot(train_acc2, label='train_acc2 (vgg13_full_lc1)')
plt.plot(val_acc2, label='val_acc2 (vgg13_full_lc1)')

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()

plt.plot(train_1_top5, label='train_1_top5 (vgg13_full)')
plt.plot(val_1_top5, label='val_1_top5 (vgg13_full)')
plt.plot(train_2_top5, label='train_2_top5 (vgg13_full_lc1)')
plt.plot(val_2_top5, label='val_2_top5 (vgg13_full_lc1)')

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()

print("vgg13 full max val acc: ", np.max(val_acc1))
print("vgg13 full lc1 max val acc: ", np.max(val_acc2))

print("vgg13 full max val top5: ", np.max(val_1_top5))
print("vgg13 full lc1 max val top5: ", np.max(val_2_top5))
```
